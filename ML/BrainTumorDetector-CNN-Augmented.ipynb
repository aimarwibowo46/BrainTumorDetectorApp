```python
import os
import cv2
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.image as mping
import tensorflow as tf
from PIL import Image
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import normalize
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Sequential, load_model
from keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Activation, Dropout, Flatten, Dense
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
# The set.seed function is used to specify the seed for the random number generator so that randomness in something can be controlled. This ensures that between the first trial and subsequent ones, the same random data is used.
np.random.seed(1234)
tf.random.set_seed(1234)
tf.keras.utils.set_random_seed(1234)
tf.config.experimental.enable_op_determinism()
```

## Load Data

```python
image_dir = 'datasets/'
no_tumor_images  = os.listdir(image_dir+ '/no')
yes_tumor_images = os.listdir(image_dir+ '/yes')
print('The length of no tumor images is',len(no_tumor_images))
print('The length of yes tumor images is',len(yes_tumor_images))
```

```python
fname1 = [os.path.join(image_dir+'/yes',fname) for fname in yes_tumor_images[0:4]]
fname2 = [os.path.join(image_dir+'/no',fname) for fname in no_tumor_images[0:4]]

ncolu = 4
nrow  = 2
fig = plt.gcf()
fig.set_size_inches(ncolu*4,nrow*4)
for i,fname in enumerate(fname1):
    sp = plt.subplot(2,4,i+1)
    sp.axis('off')
    img = mping.imread(fname)
    plt.title('Tumor Detected')
    # plt.title(f'{fname}')
    plt.imshow(img)
plt.show()

ncolu = 4
nrow = 2
fig = plt.gcf()
fig.set_size_inches(ncolu*4,nrow*4)
for i,fname in enumerate(fname2):
    sp = plt.subplot(2,4,i+1)
    sp.axis('off')
    img = mping.imread(fname)

    # Convert the image to grayscale if it's not already
    if len(img.shape) == 3 and img.shape[2] == 3:
        img = mping.imread(fname)[:, :, 0]  # Take only the first channel (red channel)
    
    plt.title('No Tumor')
    # plt.title(f'{fname}')
    # plt.imshow(img)
    plt.imshow(img, cmap='gray')  # Use 'gray' colormap
    
plt.show()
```

```python
# Create a dataframe to store file paths and labels
data = []
for label in ['yes', 'no']:  # Assuming 'yes' is for tumors and 'no' is for no tumors
    label_dir = os.path.join(image_dir, label)
    label_files = [os.path.join(label_dir, filename) for filename in os.listdir(label_dir)]
    data.extend([(filepath, label) for filepath in label_files])
    
df = pd.DataFrame(data, columns=['filepath', 'label'])
df
```

```python
# Split the data into training and testing sets
train_files, testing_files = train_test_split(df, test_size=0.2, random_state=42)

# Split the data into training and validation sets
train_files, validation_files = train_test_split(train_files, test_size=0.2, random_state=42)
```

```python
from keras.preprocessing.image import ImageDataGenerator

# Image data generators with augmentation
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255,
    rotation_range=15,
#     width_shift_range=0.1,
#     height_shift_range=0.1,
#     shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
#     fill_mode='nearest'
)

valid_datagen = ImageDataGenerator(rescale=1.0 / 255)

# Define batch size and target image size
batch_size = 32
image_size = (64, 64)

# Create data generators for training and validation
train_generator = train_datagen.flow_from_dataframe(
    train_files,
    x_col='filepath',
    y_col='label',
    target_size=image_size,
    batch_size=batch_size,
    shuffle=False,
    class_mode='binary'  # 'binary' for two classes: tumor and no tumor
)

valid_generator = valid_datagen.flow_from_dataframe(
    validation_files,
    x_col='filepath',
    y_col='label',
    target_size=image_size,
    batch_size=batch_size,
    shuffle=False,
    class_mode='binary'
)
test_generator = valid_datagen.flow_from_dataframe(
    testing_files,
    x_col='filepath',
    y_col='label',
    target_size=image_size,
    batch_size=batch_size,
    shuffle=False,
    class_mode='binary'
)
```

```python
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 5, figsize=(10,10))
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
augmented_images = [train_generator[0][0][0] for i in range(5)]
print('Augmented data result:')
plotImages(augmented_images)
```

## Building CNN Model

```python
callbacks = [ModelCheckpoint('best_model.keras', monitor='val_accuracy', verbose=1, save_best_only=True),
              EarlyStopping(monitor='val_accuracy', patience=5, verbose=1)] # Stop training when val_accuracy does not improve for 5 epochs

# patience: the number of epochs with no improvement after which the learning rate will be reduced.
# EarlyStopping: stops training when the monitored metric stops improving.
# ModelCheckpoint: saves the model or weights at certain intervals, so the model or weights can be loaded later to resume training from the saved state.
# save_best_only=True: only saves when the model is considered "best," and the latest best model according to the monitored quantity will not be overwritten.
```

```python
num_layer = [1, 2, 3]
units     = [64, 128, 256]

unit_      = []
acc_       = []
layer      = []
for x in num_layer:
    model = Sequential()
    model.add(Conv2D(32, (3,3), activation ='relu', input_shape=(64, 64, 3)))
    model.add(MaxPooling2D(2,2))
    units_ = []
    for i in range(x):
        units_.append(units[i])
        model.add(Conv2D(units[i], (3,3), activation ='relu'))
        model.add(MaxPooling2D(2,2))
    model.add(Flatten())
    model.add(Dense(64, activation = 'relu'))
    model.add(Dropout(0.5))
    model.add(Dense(1, activation = 'sigmoid'))

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    # Train the model
    epochs = 100  # You can adjust the number of epochs
    history = model.fit(
        train_generator,
        steps_per_epoch=len(train_files) // batch_size,
        epochs=epochs,
        validation_data=valid_generator,
        validation_steps=len(validation_files) // batch_size,
        callbacks = callbacks
    )
    test_loss, test_accuracy = model.evaluate(valid_generator)
    print(f"Test Accuracy: {round(test_accuracy*100, 2)}%")

    # list all data in history
    print(history.history.keys())
    # summarize history for accuracy
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'validation'], loc='upper left')
    plt.show()

    # summarize history for loss
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'validation'], loc='upper left')
    plt.show()

    acc_.append(round(test_accuracy*100, 2))
    layer.append(x)
    unit_.append(units_)
    print('-----------------------------------------------------------------------')
```

```python
# Displaying the accuracy results for each combination of the number of layers and their respective units.
df_param = pd.DataFrame({'Number of CNN Layers': layer, 'Unit': unit_, 'Accuracy(%)': acc_})
# df_param = df_param.sort_values(by=['akurasi'], ascending=False).reset_index(drop=True)
df_param
```

## Model Evaluate

```python
n_model = load_model('best_model.keras')
```

```python
y_pred = n_model.predict(test_generator)
y_pred = (y_pred > 0.5).astype(int)
```

```python
# Calculate confusion matrix
conf_matrix = confusion_matrix(test_generator.classes, y_pred)
TN = conf_matrix[0, 0]
FP = conf_matrix[0, 1]
FN = conf_matrix[1, 0]
TP = conf_matrix[1, 1]

print(f'True Positive : {TP}')
print(f'True Negative : {TN}')
print(f'False Positive : {FP}')
print(f'False Negative : {FN}')
```

```python
cm           = confusion_matrix(test_generator.classes, y_pred, labels=[1, 0])
class_label  = ['Yes', 'No']
df_confusion = pd.DataFrame(cm, index = class_label, columns = class_label)

sns.heatmap(df_confusion, annot=True, fmt = "d", cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.xlabel('Prediction Label')
plt.ylabel('Actual Label')
plt.show()
```

```python
_accuracy   = round(accuracy_score(test_generator.classes, y_pred)*100, 2)
_precission = round(precision_score(test_generator.classes, y_pred)*100, 2)
_recall     = round(recall_score(test_generator.classes, y_pred)*100, 2)
_fscore     = round(f1_score(test_generator.classes, y_pred)*100, 2)
_specificity = round(TN / (TN + FP) * 100, 2)

print('Accuracy   :', _accuracy, '%')
print('Precission :', _precission, '%')
print('Recall     :', _recall, '%')
print('F-Score    :', _fscore, '%')
print('Specificity:', _specificity, '%')
```
